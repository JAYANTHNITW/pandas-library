# Merge() Function

```python
df1=pd.DataFrame({'Roll.No':[1,2,3,4,5], 'Physics':[56,67,76,89,54]})
df1
```

```python
df2=pd.DataFrame({'Roll.No':[1,2,3,4,5], 'chemistry':[59,37,76,69,54]})
df2
```

```python
pd.merge(df1,df2, on = 'Roll.No')
```

```python
pd.merge(df1,df2) # on - by default intersecting column is taken
```

```python
df3=pd.DataFrame({'Roll.No':[1,2,3,7,8], 'Physics':[56,67,76,89,54]})
df2=pd.DataFrame({'Roll.No':[1,2,3,4,5], 'chemistry':[59,37,76,69,54]})
```

```python
pd.merge(df2, df3) # only intersecting values are merged
```

```python
pd.merge(df2, df3, how = 'left') # here we want to print  all the roll.no present in df2 irreseptive of they have value in df3.
```

```python
pd.merge(df2, df3, how = 'right')
```

```python
pd.merge(df2, df3, how = 'outer')
```

# append() Function

```python
df1=pd.DataFrame({'Roll.No':[1,2,3,4,5], 'Maths':[34,56,65,76,78], 'Physics':[56,67,76,89,54]})
df2=pd.DataFrame({'Roll.No':[6,7,8,9,10], 'Maths':[23,45,41,73,85], 'Physics':[58,27,86,89,54]})
print(df1)
print(df2)
```

```python
df1.append(df2) # Here df2 is printed after df1
```

```python
df2.append(df1)
```

```python
df1.append(df2, ignore_index=True)# now index is started from 0 and ended with 10.
```

```python
df1.append(df2, ignore_index=True,sort=True)
```

```python
df1=pd.DataFrame({'Roll.No':[1,2,3,4,5], 'chemistry':[34,56,65,76,78], 'Physics':[56,67,76,89,54]})
df2=pd.DataFrame({'Roll.No':[6,7,8,9,10], 'Maths':[23,45,41,73,85], 'Physics':[58,27,86,89,54]})
print(df1)
print(df2)
```

```python
df1.append(df2,ignore_index=True)
```
